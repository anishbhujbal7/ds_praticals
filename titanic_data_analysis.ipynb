## Data visualization

### Subtask:
Visualize the relationships between key variables in the Titanic dataset.


**Reasoning**:
Create the visualizations specified in the instructions, including a bar plot of survival rate by passenger class and gender, histograms or box plots for 'Age' and 'Fare', and potentially additional visualizations like survival rate by embarkation port or the relationship between age and survival rate.  This will be done in a single code block for efficiency.



```python
import matplotlib.pyplot as plt

# 1. Survival rate by passenger class and gender
plt.figure(figsize=(10, 6))
titanic_df.groupby(['Pclass', 'Sex'])['Survived'].mean().unstack().plot(kind='bar', ax=plt.gca())
plt.title('Survival Rate by Passenger Class and Gender')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.xticks(rotation=0)
plt.legend(title='Sex')
plt.tight_layout()
plt.show()


# 2. Histograms/box plots for 'Age' and 'Fare'
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(titanic_df['Age'].dropna(), bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.boxplot(titanic_df['Fare'].dropna(), vert=False, patch_artist=True, boxprops=dict(facecolor='lightcoral'))
plt.title('Distribution of Fare')
plt.xlabel('Fare')

plt.tight_layout()
plt.show()


```
