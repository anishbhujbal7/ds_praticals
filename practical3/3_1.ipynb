```python
file_path='/content/drive/MyDrive/colab_datasets/practical_datasets/practical3.1.csv'
```

```python
import pandas as pd

df=pd.read_csv(file_path)
df.head(10)
df.describe()
df.isnull().sum()
```

```python
df.describe()
```

```python
summary = df.groupby('Pclass')['Age'].agg(['mean', 'median', 'min', 'max', 'std'])
print(summary)
```

```python
age_list = df.dropna(subset=['Pclass', 'Age']).groupby('Pclass')['Age'].apply(list).reset_index()
print("\nList of ages for each class:")
print(age_list)
```
